```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/shristi/Downloads/Project2/sentiment_5_class.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.Phrase.tolist()
```

```python
y= df.Sentiment.tolist()
```

```python
train_set = pd.read_csv('/home/shristi/Downloads/Project2/sentiment_5_class_train.csv')
test_set = pd.read_csv('/home/shristi/Downloads/Project2/sentiment_5_class_test.csv')
X_train = train_set.Phrase.tolist()
y_train = train_set.Sentiment.tolist()

X_test = test_set.Phrase.tolist()
y_test = test_set.Sentiment.tolist()

print(len(train_set))
print(len(test_set))
```

### Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
c_vectorizer = CountVectorizer()
```

```python
c_vectorizer.fit(X_train)
c_vectorizer.get_feature_names()
```

```python
c_train_v=c_vectorizer.transform(X_train)

c_train_v.toarray()
```

```python
c_test_v=c_vectorizer.transform(X_test)
c_test_v.toarray()
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier(random_state=1)
model.fit(c_train_v,y_train)
preds=model.predict(c_test_v)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, preds)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, preds))
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_params = {
    'max_depth': (5,10,20,50, 100, 500),
    'min_samples_split': (2,3, 4,6,8,16,20)
}
```

```python
from sklearn.metrics import make_scorer, f1_score
```

```python
scorer = make_scorer(f1_score, average='micro')
clf=GridSearchCV(DecisionTreeClassifier(),grid_params,scoring=scorer,cv=3)
clf.fit(c_train_v, y_train)
```

```python
print(clf.best_score_, clf.best_params_)
```

#### Using Best parameters Decision Tree Model

```python
import sklearn.metrics as metrics
model =DecisionTreeClassifier(random_state=1,max_depth=500,min_samples_split=6)
model.fit(c_train_v, y_train)
preds = model.predict(c_test_v)
print(metrics.classification_report(y_test, preds))
```
